```python
numerical_features = ['average_forever', 'average_2weeks', 'median_forever', 'median_2weeks',
                      'ccu', 'price', 'ratings', 'owners_average']

# Function to calculate and print quartile analysis
def analyze_distribution(df, features):
    for feature in features:
        print(f"\nAnalyzing: {feature}")
        
        # Calculate quartiles
        q1 = df[feature].quantile(0.25)
        q2 = df[feature].median()
        q3 = df[feature].quantile(0.75)
        
        # Calculate IQR and outlier thresholds
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # Determine the number of data points in each quartile
        first_quartile = ((df[feature] >= lower_bound) & (df[feature] < q1)).sum()
        second_quartile = ((df[feature] >= q1) & (df[feature] < q2)).sum()
        third_quartile = ((df[feature] >= q2) & (df[feature] < q3)).sum()
        fourth_quartile = ((df[feature] >= q3) & (df[feature] <= upper_bound)).sum()
        
        # Count outliers
        outliers = ((df[feature] < lower_bound) | (df[feature] > upper_bound)).sum()
        
        # Print results
        print(f"First Quartile (Q1): {first_quartile}")
        print(f"Second Quartile (Q2): {second_quartile}")
        print(f"Third Quartile (Q3): {third_quartile}")
        print(f"Fourth Quartile (Above Q3 to upper limit): {fourth_quartile}")
        print(f"Outliers: {outliers}")

# Call the function with the desired features
analyze_distribution(df, numerical_features)
```
