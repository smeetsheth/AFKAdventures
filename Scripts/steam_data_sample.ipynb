```python
import requests

# Function to fetch player achievements for a specific game
def fetch_player_achievements(steam_id, app_id):
    api_key = "56082D97227CEB397C13EF9FB4DCA9C6"  # Replace "YOUR_ACTUAL_API_KEY" with your Steam API key
    url = f"https://api.steampowered.com/ISteamUserStats/GetPlayerAchievements/v1/?key={api_key}&steamid={steam_id}&appid={app_id}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch achievements for Steam ID {steam_id}. Status code: {response.status_code}")
        return None

# Function to fetch user stats for a specific game
def fetch_user_stats_for_game(steam_id, app_id):
    api_key = "56082D97227CEB397C13EF9FB4DCA9C6"  # Replace "YOUR_ACTUAL_API_KEY" with your Steam API key
    url = f"https://api.steampowered.com/ISteamUserStats/GetUserStatsForGame/v2/?key={api_key}&steamid={steam_id}&appid={app_id}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch user stats for Steam ID {steam_id}. Status code: {response.status_code}")
        return None
    
# Read Steam IDs from the text file
def read_steam_ids_from_file(file_path):
    with open(file_path, "r") as file:
        steam_ids = file.read().splitlines()
    return steam_ids

# Filter out public Steam IDs
def filter_public_steam_ids(steam_ids):
    public_steam_ids = []
    for steam_id in steam_ids:
        if fetch_user_attributes(steam_id) == "Public":
            public_steam_ids.append(steam_id)
    return public_steam_ids

# Example usage
file_path = "/Users/smeetsheth/Documents/steamIdList/Output.txt"  # Path to the text file containing Steam IDs
app_id = "730"  # Example app ID for the game Counter-Strike: Global Offensive

# Read Steam IDs from file
steam_ids = read_steam_ids_from_file(file_path)

# Filter public Steam IDs
public_steam_ids = filter_public_steam_ids(steam_ids)

# Open a file for writing the output
output_file_path = "record.txt"
with open(output_file_path, "w") as output_file:
    # Fetch player achievements and user stats for each public Steam ID
    for steam_id in public_steam_ids:
        achievements_data = fetch_player_achievements(steam_id, app_id)
        user_stats_data = fetch_user_stats_for_game(steam_id, app_id)
        # Write the obtained data to the output file
        output_file.write(f"Player achievements for Steam ID {steam_id}: {achievements_data}\n")
        output_file.write(f"User stats for Steam ID {steam_id}: {user_stats_data}\n")

print("Output written to", output_file_path)
```

```python
import requests

# Function to fetch public Steam IDs and save them to a local file
def fetch_and_save_public_steam_ids(api_key, num_ids=1000, output_file="public_steam_ids.txt"):
    url = f"https://api.steampowered.com/ISteamUser/GetPlayerSummaries/v2/?key={api_key}&format=json&steamids="
    public_ids = []

    for i in range(1, num_ids + 1):
        response = requests.get(url + str(i))
        if response.status_code == 200:
            data = response.json()
            if data.get("response") and data["response"].get("players"):
                for player in data["response"]["players"]:
                    if player.get("communityvisibilitystate") == 3:
                        public_ids.append(player["steamid"])
            else:
                break  # No more players found
        else:
            print(f"Failed to fetch data for batch {i}. Status code: {response.status_code}")
            break  # Stop fetching on error

    # Save public Steam IDs to a local file
    with open(output_file, "w") as file:
        for steam_id in public_ids:
            file.write(steam_id + "\n")

    return public_ids

# Example usage
api_key = "56082D97227CEB397C13EF9FB4DCA9C6"  # Replace "YOUR_ACTUAL_API_KEY" with your Steam API key
output_file = "public_steam_ids.txt"  # Specify the output file path
public_steam_ids = fetch_and_save_public_steam_ids(api_key, num_ids=1000, output_file=output_file)
print("Public Steam IDs saved to:", output_file)
```

```python
def generate_recommendations(similarity_matrix, train_data, test_data, top_n=5):
    recommendations = defaultdict(list)
    for user_id in test_data['SteamID'].unique():
        user_games = test_data[test_data['SteamID'] == user_id]['GameID']
        for game_id in user_games:
            sim_scores = similarity_matrix[train_data['GameID'] == game_id]
            sim_scores = sim_scores.flatten()
            top_sim_indices = sim_scores.argsort()[-top_n-1:-1][::-1]  # Exclude the game itself
            similar_games = train_data.iloc[top_sim_indices]['GameID'].values
            recommendations[user_id].extend(similar_games)
    return recommendations
```

```python
from sklearn.metrics import precision_score

# Assuming 'test_set' contains the test data with actual interactions
# 'recommendations' contains the recommended items for each user
# 'similarity_matrix' contains the cosine similarity scores

# Function to generate recommendations for each user
def generate_recommendations(X_test, similarity_matrix, threshold=0.5):
    recommendations = {}
    for user_id in X_test['SteamID'].unique():
        if user_id in similarity_matrix:
            # Find top similar items based on similarity scores
            sim_scores = similarity_matrix[user_id]
            similar_items = [item_id for item_id, score in sim_scores if score > threshold]
            recommendations[user_id] = similar_items
    return recommendations

# Generate recommendations for test set
recommendations = generate_recommendations(X_test, similarity_matrix)

# Function to evaluate precision
def evaluate_precision(X_test, recommendations):
    y_true = []
    y_pred = []
    for user_id in X_test['SteamID'].unique():
        actual_items = X_test[X_test['SteamID'] == user_id]['GameID'].values
        recommended_items = recommendations.get(user_id, [])
        y_true.extend(actual_items)
        y_pred.extend(recommended_items)
    return precision_score(y_true, y_pred)

# Evaluate precision
precision = evaluate_precision(X_test, recommendations)
print("Precision:", precision)




```
