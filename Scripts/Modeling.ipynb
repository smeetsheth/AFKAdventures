```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
from sklearn.preprocessing import StandardScaler

# Ensure the original DataFrame contains both 'appid' and 'name' columns
appid_name_lookup = df[['appid', 'name']].drop_duplicates()

# Aggregating user and game profiles
user_profiles = df.groupby('SteamID').agg(
    average_playtime=('PlaytimeForever_standardized', 'mean'),
    total_playtime=('PlaytimeForever_standardized', 'sum'),
    average_ratings=('ratings', 'mean'),
    preferred_genre=('genre_encoded', lambda x: x.mode()[0])  # Most common genre
).reset_index()

game_profiles = df.groupby('appid').agg(
    average_playtime=('PlaytimeForever_standardized', 'mean'),
    total_playtime=('PlaytimeForever_standardized', 'sum'),
    average_ratings=('ratings', 'mean'),
    primary_genre=('genre_encoded', lambda x: x.mode()[0])  # Most common genre
).reset_index()

# Merge with the appid-name lookup to include game names
game_profiles = game_profiles.merge(appid_name_lookup, on='appid', how='left')

# Extract common features between user and game profiles
common_features = set(user_profiles.columns) & set(game_profiles.columns)
user_features = user_profiles[list(common_features)].drop(['SteamID'], axis=1, errors='ignore')
game_features = game_profiles[list(common_features)].drop(['appid'], axis=1, errors='ignore')

# Standardize the feature data
scaler = StandardScaler()
user_features_scaled = scaler.fit_transform(user_features)
game_features_scaled = scaler.transform(game_features)

# Function to compute Pearson's correlation between two feature vectors
def pearson_correlation(vector1, vector2):
    correlation, _ = pearsonr(vector1, vector2)
    return correlation

# Function to generate top N recommendations using Pearson's correlation
def get_pearson_recommendations(user_index, user_features_scaled, game_features_scaled, game_profiles, top_n=5):
    # Ensure the user index is within bounds
    if user_index < 0 or user_index >= user_features_scaled.shape[0]:
        print("Invalid user index!")
        return

    # Extract the feature vector for the target user
    target_user_vector = user_features_scaled[user_index]

    # Calculate Pearson correlation for the target user against all game features
    similarity_scores = []
    for game_vector in game_features_scaled:
        similarity_scores.append(pearson_correlation(target_user_vector, game_vector))

    # Convert similarity scores to a pandas Series for easier ranking
    similarity_scores_series = pd.Series(similarity_scores)

    # Rank the games by similarity score (descending order)
    top_indices = similarity_scores_series.argsort()[::-1][:top_n]

    # Retrieve the top N recommended games' appids and names
    recommended_games = game_profiles.iloc[top_indices][['appid', 'name']]

    print(f"\nTop {top_n} recommended games for user {user_index}:")
    print(recommended_games)

# Ask for the user index input
try:
    target_user_index = int(input("Enter the user index (0 to {}): ".format(user_features_scaled.shape[0] - 1)))
except ValueError:
    print("Please enter a valid numeric index.")
    target_user_index = -1

# Retrieve and display recommendations if the user index is valid
if target_user_index >= 0:
    get_pearson_recommendations(target_user_index, user_features_scaled, game_features_scaled, game_profiles)
```
